```python
p = r'https://'
Dawn_news['text-3 href']=Dawn_news['text-3 href'].str.replace(p,'')
print(Dawn_news.head())
```

```python
pattern = r'[^a-zA-Z\s]'
Dawn_news['story__link']=Dawn_news['story__link'].str.replace(pattern,'')
print(Dawn_news.head())
```

```python
pattern1= r'^.*?(\d)'
Dawn_news['story__link href']=Dawn_news['story__link href'].str.replace(pattern1,r'\1')
Dawn_news['story__link href']=Dawn_news['story__link href'].str.replace('-',' ')
print(Dawn_news.head())
```

```python
Dawn_news['story__link href']=Dawn_news['story__link href'].str.replace('/','-')
print(Dawn_news.head())
```

```python
xy=r'[^\w\s\'",.]'
Dawn_news['story__excerpt']=Dawn_news['story__excerpt'].str.replace(xy,'')
print(Dawn_news.head())
```

```python
pattern2 = r'\s\d{1,2}:\d{2}[ap]m$'
Dawn_news['timestamp--time']=Dawn_news['timestamp--time'].str.replace(pattern2,'')
print(Dawn_news.head())
```

```python
#story__link   story__excerpt
Dawn_news['Concatenated_Columns']=Dawn_news['story__link']+Dawn_news['story__excerpt']
print(Dawn_news.head())
```

```python
Dawn_news.to_csv("cleardata.csv")
```

```python
pakistan_rows = Dawn_news[Dawn_news['text-3'].str.contains(r'\bpakistan\b', flags=re.IGNORECASE)]
world_rows = Dawn_news[Dawn_news['text-3'].str.contains(r'\bworld\b', flags=re.IGNORECASE)]
newspaper_rows = Dawn_news[Dawn_news['text-3'].str.contains(r'\bnewspaper\b', flags=re.IGNORECASE)]

# Save each subset into separate files
pakistan_rows.to_csv('pakistan_rows.csv', index=False)
world_rows.to_csv('world_rows.csv', index=False)
newspaper_rows.to_csv('newspaper_rows.csv', index=False)
```
